```python
# 範例：多張圖片的OCR光學文字辨識
import keras_ocr
import matplotlib.pyplot as plt
pipeline = keras_ocr.pipeline.Pipeline()
images = []
imgfiles = [             # 圖片串列：將要辨識文字的原始圖片檔案一一列出
    'ad1.jpg',
    'ad2.jpg',
]
for imgfile in imgfiles:
    images.append(keras_ocr.tools.read(imgfile))
predict = pipeline.recognize(images)
_, axs = plt.subplots(ncols=len(images), figsize=(20,20))
for i in range(len(predict)):
    keras_ocr.tools.drawAnnotations(image=images[i],
                                    predictions=predict[i],
                                    ax=axs[i])

# 多張圖片的OCR光學文字辨識時，若圖片數量越多或圖片尺寸越大，則執行時間會較長，需要耐心等待！
# 以辨識 ad1.jpg 及 ad2.jpg 為例，在 未使用GPU 情況下，耗時約 2分鐘54秒 才執行完畢。
```

```python
# 應用範例：車牌辨識 (以 1710YC_all.jpg 測試辨識車牌效果為依據)
import keras_ocr
import re

def checkcharnum(str1):
    if re.match("^[a-z0-9]*$", str1):
        return True
    else:
        return False

pipeline = keras_ocr.pipeline.Pipeline()
images = []
images.append(keras_ocr.tools.read('ABP0023_all.jpg'))
predict = pipeline.recognize(images)
result = ''
if len(predict[0]) == 1:
    result = predict[0][0][0]    # 取出辨識出的文字
else:
    txt = []
    xpos = []
    for i in range(len(predict[0])):
      temstr = predict[0][i][0]        # 圖片中辨識出的第i組文字
      if checkcharnum(temstr) and len(temstr)<=7:
          txt.append(temstr)
          xpos.append(predict[0][i][1][0][0])   # 圖片中辨識出的第i組文字在圖片中的左上角位置 (X座標值)
    xtem = xpos.copy()
    xtem.sort()   # 將辨識出的文字以其X座標值來排序，以獲得正確的車牌號碼順序
    for i in range(len(xpos)):
        result += txt[xpos.index(xtem[i])]
result = result.replace('o','0').replace('i','1').upper()  # 因新車牌已無使用 o及i ，所以將辨識出的 o及i 以  0及1 來取代。
print('車牌號碼：{}'.format(result))
```

```python
# 範例：自製口罩偵測程式
import cv2
from google.colab.patches import cv2_imshow

# 自行選擇一個圖檔來測試是否有戴口罩 (mask1.jpg、maskhalf.jpg、person1.jpg、mask2.jpg)
#img = cv2.imread('mask1.jpg')       # 有戴口罩 mask1.jpg
#img = cv2.imread('maskhalf.jpg')    # 口罩沒戴好 (只戴一半) maskhalf.jpg
#img = cv2.imread('person1.jpg')     # 沒有戴口罩 person1.jpg
img = cv2.imread('mask2.jpg')        # 1個人有戴口罩，1個人口罩沒戴好(只戴一半) mask2.jpg

# 偵測臉的模型
facemodel = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
# 偵測鼻子的模型
nosemodel = cv2.CascadeClassifier('haarcascade_mcs_nose.xml')

faces = facemodel.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30,30))
noses = nosemodel.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30,30))

for (xf,yf,wf,hf) in faces:
    flag = False
    for (xn,yn,wn,hn) in noses:
        #檢查臉範圍內是否偵測到鼻子
        if (xn>=xf and xn<=(xf+wf)) and ((xn+wn)>=xf and (xn+wn)<=(xf+wf)) and\
            (yn>=yf and yn<=(yf+hf)) and ((yn+hn)>=yf and (yn+hn)<=(yf+hf)):  #偵測到鼻子
            flag = True
            break
    if flag:  #未載口罩以「紅色」框選
        cv2.rectangle(img, (xf,yf), (xf+wf,yf+hf), (0,0,255))
    else:  #有載口罩以「綠色」框選
        cv2.rectangle(img, (xf,yf), (xf+wf,yf+hf), (0,255,0))

cv2_imshow(img)   # 將偵測結果繪製出來 (有載口罩以「綠色」框選；未載口罩以「紅色」框選)
```

## 臉部偵測

```python
# 臉部偵測 語法：
#    圖片變數 = DeepFace.detectFace(img_path=圖片路徑, detector_backend=模型名稱, enforce_detection=布林值)
# 語法中的「模型名稱」可為：opencv、retinaface、mtcnn、dlib、ssd。
#【注意：ssd會產生錯誤，所以不會將臉部切割出來回傳，而是傳回原始圖片】

image = DeepFace.detectFace(img_path='/content/person1.jpg',   # 偵測臉部的原始圖片為 person1.jpg
                           detector_backend='dlib',
                           enforce_detection=False)
plt.imshow(image)   # 執行結果只會將臉部切割回傳，而不是在原圖中框選標示出人臉部分
```

```python
image.shape
```

```python
# 將每個像素值乘上255後，再重新存檔一次(例如：bear.jpg)
image *= 255
cv2.imwrite('bear.jpg', image)

# 此時若開啟儲存起來的圖檔(例如：bear.jpg)，則可看到臉部圖案，但是色階不對 (偏藍，與原圖不圖)!
# 色階不對是因為 OpenCV模型的圖像是使用BGR格式，而非RGB格式。所以還必須將RGB格式轉為BGR格式才能顯示正確色階。
```

```python
# 將圖像的RGB格式轉為BGR格式後，再重新存檔一次(例如：bear.jpg)
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
cv2.imwrite('bear.jpg', image)

# 此時開啟儲存起來的圖檔(例如：bear.jpg)，則可看到臉部圖案，且顯示色階正確 (與原圖相同)。
```

### 應用：WebCam人臉登入系統
需要搭配使用「網路攝影機」。

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):    # 攝影機抓取擷圖將儲存為 photo.jpg
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)    # 執行顯示上面的Javascript程式

  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)

  return filename
```

```python
# 「正面情緒資料檔」 完整程式碼
all = pd.read_csv('/content/hotel_all.csv')
posall = all[all['label']==1]
pos = posall.sample(2444)          # 亂數取出2444筆資料 (為了跟「負面情緒資料檔」筆數相同)
testpos = pos.iloc[:100]           # 取出前100筆資料作為測試資料 (其他的2344筆為訓練資料)
pos = pos.drop(columns='label')    # 將'label'欄位刪除掉 (測試資料不需要'label'欄位)
postrain = pos.iloc[100:]          # 第100筆以後的資料作為訓練資料 (2344筆訓練資料)
postrain.to_csv('postrain.csv', header=False, index=False)
```

```python
# 「負面情緒資料檔」 完整程式碼
negall = all[all['label']==0]
neg = negall.sample(frac=1.0)
testneg = neg.iloc[:100]          # 取出前100筆資料作為測試資料 (其他的2344筆為訓練資料)
neg = neg.drop(columns='label')   # 將'label'欄位刪除掉 (測試資料不需要'label'欄位)
negtrain = neg.iloc[100:]         # 第100筆以後的資料作為訓練資料 (2344筆訓練資料)
negtrain.to_csv('negtrain.csv', header=False, index=False)
```

```python
# 建立測試資料檔 (合併「正面情緒資料檔」及「負面情緒資料檔」的測試資料)
testall = pd.concat([testpos, testneg], axis=0)    # axis=0 表示按照列來合併「正面情緒資料檔」及「負面情緒資料檔」的資料
testall = testall.sample(frac=1.0)                 # 將資料順序打亂，避免相同情緒的資料集中在一起
testall.to_csv('testall.csv', header=False, index=False)     # 儲存測試資料檔為 testall.csv
```

```python
# 先使用snownlp模組內建的情緒分析模型來進行情緒分析 (尚未訓練)，看看正確率如何？

score = 0
with open("testall.csv", "r") as f:
    datas = f.readlines()
    for data in datas:
        label = data.split(',')[0]
        text = data.split(',')[1]
        if SnowNLP(text).sentiments<0.5:
            ss = 0
        else:
            ss = 1
        if int(label) == ss:
            score +=1
print("snownlp模組內建情緒分析模型的正確率：{}".format(score/len(datas)))
```

## 【數學與時間模式】ChatBot
可回覆簡單的數學運算式結果，或是時間。

```python
# 數學與時間模式【建立機器人物件】語法：
#     機器人物件變數 = ChatBot(機器人名稱, storage_adapter=資料庫轉接器, logic_adapters=邏輯轉接器)
# 語法中的「邏輯轉接器」可為【數學與時間模式】語法：
#     logic_adapters=[ 'chatterbot.logic.MathematicalEvaluation', 'chatterbot.logic.TimeLogicAdapter']

bot = ChatBot(
    'MathTimeBot',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    logic_adapters=[
        'chatterbot.logic.MathematicalEvaluation',     # 此為負責數學的回覆
        'chatterbot.logic.TimeLogicAdapter'            # 此為負責時間的回覆
    ]
)
```

## 【串列訓練模式】ChatBot
給連續的問句及答句(Q&A)，當提出問句時，會回覆其答句。

```python
# 串列訓練模式【建立機器人物件】語法：
#     機器人物件變數 = ChatBot(機器人名稱, storage_adapter=資料庫轉接器, logic_adapters=邏輯轉接器)
# 語法中的「邏輯轉接器」可為【串列訓練模式】語法：
#     logic_adapters=[{ 'import_path': 'chatterbot.logic.BestMatch', 'default_response': 預設回應文句, 'maximum_similarity_threshold': 數值, }]

bot = ChatBot(
    'SimpleBot',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    logic_adapters=[
        { 'import_path': 'chatterbot.logic.BestMatch',
          'default_response': '很抱歉！我不了解你的意思。',
          'maximum_similarity_threshold': 0.6,    # 問題相似度高於threshold值 (例如 0.6) 才會回覆相對應的答句
        }
    ]
)

trainer = ListTrainer(bot)
trainer.train([
    '你好',                  # 問句1
    '你好',                  # 答句1
    '有什麼能幫你的？',       # 問句2
    '想買資料科學的課程',     # 答句2
    '具體是資料科學哪塊呢？'  # 問句3
    '機器學習',              # 答句3
])
```

```python
# 向【串列訓練模式】ChatBot 提出問題，看其回覆狀況。
# 問題相似度高於threshold值 (例如 0.6) 才會回覆相對應的答句
q = '什麼能幫你'
#q = '有什麼能幫你的？'

response = bot.get_response(q)
response
```

```python
# 建立chatterbot機器人物件
bot = ChatBot(
    'SimpleBot',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    logic_adapters=[
        { 'import_path': 'chatterbot.logic.BestMatch',
          'default_response': '很抱歉！我不了解你的意思。',
          'maximum_similarity_threshold': 0.6,
        }
    ]
)
```

```python
# 向【以自訂資料建立模型】ChatBot提出問題，看其回覆狀況。
q = '看診時間'
response = bot.get_response(q)
response
```
